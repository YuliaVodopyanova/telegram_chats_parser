Now, combining all the files

```python
combined_data_2_chats_new = pd.concat([df_new, df_1st_chat], ignore_index=True)
```

```python
combined_data = pd.concat([combined_data_2_chats_new, previously_used_df], ignore_index=True)
```

As the dates mights overlap, it's logical to remove the duplicates

```python
combined_data = combined_data.drop_duplicates()
```

```python
len(combined_data)
```

and delete the empty 'lemmas' rows

```python
combined_data.isna().sum()
```

```python
combined_data = combined_data.dropna(subset=['Lemmas'])
```
